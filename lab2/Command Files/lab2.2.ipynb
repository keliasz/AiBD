```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Example plots

```python
# read csv analysed data
df_tb = pd.read_csv('../Analysis Data/tb_analysed.csv')
```

```python
df_tb.head()
```

```python
n_tb_sum = df_tb.groupby('year')['cases'].sum()
n_tb_sum_2 = pd.DataFrame(n_tb_sum)
# example plot data x - year, y - cases
plt.rcParams["figure.figsize"] = (20,10)
n_tb_sum_2.plot(kind='bar')
plt.ylabel('cases')
plt.grid()
plt.yscale('log')
plt.show()
```

```python
n_tb_age_sum = df_tb.groupby(['age','sex'])['cases'].sum()
n_tb_age_sum_2 = pd.DataFrame(n_tb_age_sum)
# example plot data x - (age, sex), y - cases
n_tb_age_sum_2.plot(kind='bar', rot=0)
plt.ylabel('cases')
plt.grid()
plt.show()
```
