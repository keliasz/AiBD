# Zbiór danych numer 11 - województwo śląskie

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
slaskie_df = pd.read_csv('../Original Data/11_S╠üLA╠ĘSKIE.csv', index_col=0)
print(len(slaskie_df))
```

```python
slaskie_df.head()
```

```python
slaskie_df.info()
```

```python
ind = slaskie_df.loc[slaskie_df['Płeć kupującego']=='bd.'].index
slaskie_dropna_plec = slaskie_df.drop(ind)
```

```python
sns.set_theme(style='darkgrid')
sns.countplot(x="Dni od zakupu", data=slaskie_dropna_plec)
plt.show()
```

```python
sns.countplot(x="Marka", data=slaskie_df)
plt.show()
```

```python
bins_edges = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
slaskie_df['Wiek kupującego'].hist(bins=bins_edges)
plt.xlabel('Wiek kupującego')
plt.ylabel('count')
plt.show()
```

```python
sns.countplot(x="Płeć kupującego", data=slaskie_dropna_plec)
plt.show()
```

```python
sns.countplot(x='Ocena', data=slaskie_df)
plt.show()
```

```python
sns.countplot(x='Ocena',hue='Płeć kupującego', data=slaskie_dropna_plec)
plt.legend(loc='upper right')
plt.show()
```

```python
f=plt.figure()
sns.countplot(x='Marka',hue='Ocena', data=slaskie_df)
plt.legend(loc='upper right')
f.set_figwidth(15)
f.set_figheight(10)
plt.show()
```
